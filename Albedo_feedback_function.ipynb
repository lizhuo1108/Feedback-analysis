```python
def alb_feedback(dalb):
    #---------------read the kernel------------------------
    diri='/zhoulab_rit/lzhuo/data/Pendergrass_kernel/'
    filename='alb.kernel.nc'
    fi=xr.open_dataset(diri+filename)
    alb_kernel=fi.FSNT
    alb_kernel=preprocess_2D(alb_kernel)
    alb_kernel_cs=fi.FSNTC
    alb_kernel_cs=preprocess_2D(alb_kernel_cs)
    #--------------regrid dalb onto the coordinates of the kernel--
    new_lat=alb_kernel.lat
    new_lon=alb_kernel.lon
    dalb=dalb.interp(lat=new_lat,lon=new_lon,kwargs={"fill_value": "extrapolate"})
    #--------------do the calculation--------------------
    dSW_alb=albedo_feedback(alb_kernel,dalb)
    dSW_alb_cs=albedo_feedback(alb_kernel_cs,dalb)
    return dSW_alb,dSW_alb_cs
```
